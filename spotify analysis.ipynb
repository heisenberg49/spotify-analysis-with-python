```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv(r'C:\Users\prem\Downloads\Spotify dataset\tracks.csv')
df_tracks.head()
```

```python
# checking null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# top 10 least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
#top 10 most popular songs
sorted_df = df_tracks.sort_values('popularity', ascending = False)
sorted_df[:10]
```

```python
most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending = False)
most_popular[:10]
```

```python
df_tracks.describe().transpose()
```

```python
df_tracks.set_index('release date', inplace = True)
df_tracks.head()
```

```python
#artist name in 25th row
df_tracks[['artists']].iloc[25]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis =1)
```

```python
df_tracks.head()
```

```python
#co-relation
corr_df = df_tracks.drop(['key','mode','explicit'],axis = 1).corr(method = 'pearson')
plt.figure(figsize=(15,7))
heatmap = sns.heatmap(corr_df,annot = True,fmt ='.2g',vmin=-1,vmax=1,center = 0,cmap = 'Blues',linewidth=1,linecolor = 'black')
heatmap.set_title('correlation heatmap between variables')
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x= 'energy',y = 'loudness',color='red').set(title ='Energy vs Loudness Regression')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x= 'acousticness',y = 'popularity',color='blue').set(title ='Acousticness vs Popularity Regression')
```
